```python
def conv_to_timestamp(seconds):
    total_secs = datetime.timedelta(seconds=seconds)
    mins, secs = divmod(total_secs.seconds, 60)
    hours, _ = divmod(mins, 60)
    millis = total_secs.microseconds/1000
    millis = str(int(millis))[:3]
    while len(millis) < 3:
        millis += "0"
    timestamp_str = "{:02d}:{:02d}:{:02d}.{}".format(hours,mins,secs,millis)
    return timestamp_str
```

```python
conv_to_timestamp(result['segments'][0]['start'])
```

```python
vtt = WebVTT()

for line in result['segments']:
    caption = Caption(
        conv_to_timestamp(line['start']),
        conv_to_timestamp(line['end']),
        line['text']
    )
    vtt.captions.append(caption)
```
